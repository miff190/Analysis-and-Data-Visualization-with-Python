## Dasar Pie Chart

```python
import matplotlib.pyplot as plt
Data1 = [25,35,15,5,20]
fig = plt.figure(figsize=(10,5))

plt.pie(Data1)
plt.title('Penjualan Sepeda')
plt.show()
```

## Label dan Persen

```python
# Membuat label
import matplotlib.pyplot as plt
Merek = ['A','B','C','D','E'] # <-- buat list label
Data1 = [25,35,15,5,20]
fig = plt.figure(figsize=(10,5))

plt.pie(Data1, labels=Merek) # <-- terapkan label di sini
plt.title('Penjualan Sepeda')
plt.show()
```

```python
# Membuat persentase
plt.pie(Data1, labels=Merek, autopct='%1.2f%%') # <-- autopct='%1.2f' berarti ada dua angka di belakag koma
plt.title('Penjualan Sepeda')
plt.show()
```

## Parameter startangle
agar potongan pie pertama bisa tegak lurus

```python
# Membuat persentase
plt.pie(Data1, labels=Merek, autopct='%1.2f%%', startangle=90) # <-- potongan pie pertama tegak lurus 90 derajat
plt.title('Penjualan Sepeda')
plt.show()
```

## Parameter explode
agar potongan pie keluar dari gabungan

```python
# Membuat label
Merek = ['A','B','C','D','E']
Data1 = [25,35,15,5,20]
Explode = [0,0.1,0,0,0] # <-- menjauh 0.1

plt.pie(Data1,
        explode=Explode, # <-- terapkan di sini
        labels=Merek,
        autopct='%1.2f%%',
        startangle=90)
plt.title('Penjualan Sepeda')
plt.show()
```

## Pie Chart dengan Plotly

```python
import plotly.express as px
import pandas as pd

Data = {'Merek':['Honda','Toyota','Daihatsu','Yamaha','BMW'],
        'Penjualan':[25,35,15,5,20]}

df_Data = pd.DataFrame(Data)
chart = px.pie(df_Data, values='Penjualan', names='Merek')
chart.show()
```
