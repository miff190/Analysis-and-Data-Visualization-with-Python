## Line Style, Marker, dan Color

### 1. Parameter ls

```python
# library
import matplotlib.pyplot as plt

# data
Bulan = ['Jan','Feb','Mar','Apr','Mei','Jun']
Data1 = [45,67,30,22,90,100]
Data2 = [50,69,55,30,75,90]

# design
fig = plt.figure(figsize=(10,5))
plt.title('Data Penjualan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan dalam Jutaan')

# plotting
chart = plt.subplot()
chart.plot(Bulan, Data1, ls=':') # <-- terapkan di sini
chart.plot(Data2, ls='dashdot')
plt.show()

# keterangan :
# ls='-' solid
# ls='--' dashed
# ls='-.' dashdot
# ls=':' dotted
```

### 2. Parameter marker

```python
# library
import matplotlib.pyplot as plt

# data
Bulan = ['Jan','Feb','Mar','Apr','Mei','Jun']
Data1 = [45,67,30,22,90,100]
Data2 = [50,69,55,30,75,90]

# design
fig = plt.figure(figsize=(10,5))
plt.title('Data Penjualan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan dalam Jutaan')

# plotting
chart = plt.subplot()
chart.plot(Bulan, Data1, ls=':', marker='o') # <-- terapkan di sini
chart.plot(Data2, ls='--', marker='<')
plt.show()

# keterangan :
# marker='.' point
# marker=',' pixel
# marker='o' circle
# marker='v' triangle_down
# marker='^' triangle_up
# marker='<' triangle_left
# marker='>' triangle_right
# marker='1' tri_down
# marker='2' tri_up
```

### Parameter color

```python
# library
import matplotlib.pyplot as plt

# data
Bulan = ['Jan','Feb','Mar','Apr','Mei','Jun']
Data1 = [45,67,30,22,90,100]
Data2 = [50,69,55,30,75,90]

# design
fig = plt.figure(figsize=(10,5))
plt.title('Data Penjualan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan dalam Jutaan')

# plotting
chart = plt.subplot()
chart.plot(Bulan, Data1, ls=':', marker='o', color='g') # <-- terapkan di sini
chart.plot(Data2, ls='--', marker='<', color='r')
plt.show()

# keterangan :
# color='b' blue
# color='g' green
# color='r' red
# color='c' cyan
# color='m' magenta
# color='y' yellow
# color='k' black
# color='w' white
```

## Multiple Subplot
menampilkan beberapa line plot

```python
# persiapan data
import matplotlib.pyplot as plt
Bulan = ['Jan','Feb','Mar','Apr','Mei','Jun']
Data1 = [45,67,30,22,90,100]
Data2 = [50,69,55,30,75,90]
fig = plt.figure(figsize=(10,10))

# buat plot
chart1 = plt.subplot(311)
chart2 = plt.subplot(312)

# label
plt.xlabel('Bulan')
plt.ylabel('Penjualan dalam Jutaan')

chart1.set(title='Data Penjualan A')
chart2.set(title='Data Penjualan B')

# design
chart1.plot(Bulan, Data1, ls=':', marker='o', color='g')
chart2.plot(Bulan, Data2, ls='--', marker='<', color='r')

plt.show()
```

## Mathod .savefig()
untuk menyimpan plot ke ekstensi gambar

```python
import matplotlib.pyplot as plt
Bulan = ['Jan','Feb','Mar','Apr','Mei','Jun']
Data1 = [45,67,30,22,90,100]
Data2 = [50,69,55,30,75,90]
fig = plt.figure(figsize=(10,10))

chart1 = plt.subplot(311)
chart2 = plt.subplot(312)

plt.xlabel('Bulan')
plt.ylabel('Penjualan dalam Jutaan')

chart1.set(title='Data Penjualan A')
chart2.set(title='Data Penjualan B')

chart1.plot(Bulan, Data1, ls=':', marker='o', color='g')
chart2.plot(Bulan, Data2, ls='--', marker='<', color='r')

plt.savefig('lineplot') # <-- nama gambar; default .PNG
plt.show()
```

## Line Chart dengan Plotly
alternatif selain menggunakan matplotlib

```python
import plotly.express as px

Nomor = [1,2,3,4,5]
DataPenjualan = [300,200,150,500,700]

# width dan height satuannya px
chart = px.line(x=Nomor, y=DataPenjualan,
                title='Data Penjualan',
                width=800,
                height=400)
chart.show()
```

```python
# terapkan dataset csv ke plot
import pandas as pd
import plotly.express as px

Data = {'Nomor':[1,2,3,4,5],
       'DataPenjualan':[300,200,150,500,700]
        }
df_Data = pd.DataFrame(Data)

chart = px.line(df_Data, x='Nomor', y='DataPenjualan',
                title='Data Penjualan',
                width=800,
                height=400)
chart.show()
```

## Method .update_layout()
mengatur penampang di luar chart

```python
# manfaatkan paper_bgcolor untuk warna latar belakang
chart.update_layout(paper_bgcolor='LightSteelBlue')
chart.show()
```

```python
# gunakan parameter margin untuk mengatur jarak gambar dengan background
chart.update_layout(paper_bgcolor='LightSteelBlue',
                    margin=dict(l=50, r=50, t=100, b=20))
chart.show()
```
