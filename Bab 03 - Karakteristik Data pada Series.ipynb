## List dan Dictionary

```python
import pandas as pd
```

```python
# index selalu dimulai dengan 0, kalo list indeksnya gabisa diubah
teman = ['Joni Gudel', 'Edi Yunani', 'Toni Sulung', 'Retno Mimin']
pd.Series(teman)
```

```python
# mengatur label pada dictionary, indeksnya bisa diubah
temanakrab = {3:'Joni Gudel', 2:'Edi Yunani', 1:'Toni Sulung', 0:'Retno Mimin'}
pd.Series(temanakrab)
```

```python
# parameter index
pd.Series(data=teman, index=['teman1','teman2','teman3','teman4'])
```

## Atribut Name

```python
namateman = pd.Series(teman)
namateman.name = 'Daftar Nama Teman Akrab'
namateman
```

## Method Head() dan Tail()

```python
# membuat series dari angka 0 sampai 99
angka = pd.Series(range(0,100)) # angka=variabel; .Series()=konstruktor; range()=method; 0=start; 100=stop
angka
```

```python
# menampilkan 5 data teratas
angka.head()
```

```python
# menampilkan 10 data terbawah
angka.tail(10)
```
