## Import Dataset

```python
import pandas as pd

# akses data melalui url
data_url = 'https://thinkjubilee.com/bola.csv'
pd.read_csv(data_url)
```

## Boolean Mask

```python
pemainbola = pd.read_csv('https://thinkjubilee.com/bola.csv')
pemainbola['market_value'] > 50
```

```python
# tampilan lebih lengkap
pemainbola[pemainbola['market_value'] > 50]
```

```python
# tampilan lebih ringkas
pemainbola[pemainbola['market_value'] > 50].shape
```

## Method unique()
mengambil data unik gak ada yang sama

```python
# method unique
pemainbola['nationality'].unique()
```

```python
# mengetahui jumlah negara yang unik
pemainbola['nationality'].unique().size
```

## Method isin()
untuk mengetahui data-data yang berada pada kriteria tertentu

```python
# tampilan bolean mask
pemainbola['nationality'].isin(['Englang','South Korea','Italy'])
```

```python
# tampilan secara lengkap
pemainbola[pemainbola['nationality'].isin(['Englang','South Korea','Italy'])]
```

## Method between()
mendapatkan range nilai dari dataset. Butuh 2 argumen --> nilai terendah & nilai tertinggi

```python
pemainbola[pemainbola['market_value'].between(65,75)]
```

## Operator dan Method Perbanndingan
digunakan untuk memfilter dataset. operatornya adalah ==,>,<, dsb

```python
pemainbola = pd.read_csv('https://thinkjubilee.com/bola.csv')

# para pemain dengan usia dibawah 30 tahun
pemainbola[pemainbola['age'] < 30]
```

```python
# kurang dari (<) setara dengan method lt()
pemainbola[pemainbola['age'].lt(30)]

# le() setara dengan <=
# lt() setara dengan <
# gt() setara dengan >
# ge() setara dengan >=
# eq() setara dengan ==
```

```python
# kombinasi lebih 2 kondisi
pemainbola[(pemainbola['market_value'] > 50) & (pemainbola['nationality'] == 'England')]
```

```python
# buat lebih spesifik lagi
pemainbola[
    (pemainbola['market_value'] > 50) &
    (pemainbola['nationality'] == 'England') &
    (pemainbola['age'] < 30)]
```

```python
# manfaatkan variabel biar gak terlalu pusing
usia = pemainbola['age'] < 30
klub = pemainbola['club'] == 'Arsenal'
market = pemainbola['market_value'] > 30

# terapkan ke penampil
pemainbola[usia & klub & market]
```

```python
# gunakan operator NOT (~)
pemainbola[usia & ~klub & market].head()
```

## Mengenal 2D Indexing
indeksing menurut baris dan kolom sekaligus

```python
# indexing 2D
pemainbola.loc[usia & ~klub & market,
              ['name','club','nationality','market_value']]
```
