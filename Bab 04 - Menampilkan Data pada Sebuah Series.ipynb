## Menampilkan Nilai

```python
import pandas as pd

teman = ['Joni Gudel', 'Edi Yunani', 'Toni Sulung', 'Retno Mimin']

# menampilkan teman indeks ke-1
print(teman[1])
```

```python
# menampilkan teman indeks ke-1 sampai 3 menggunakan range
print(teman[1:3])
```

```python
# dengan parameter index
d = pd.Series(data=teman, index=['teman1','teman2','teman3','teman4'])
print(d['teman1'])
```

```python
# menampilkan teman ke-2 sampai 3 menggunakan range di parameter indeks
print(d['teman2':'teman3'])
```

## Properti .loc

```python
a = pd.Series(teman)
a.loc[0]
```

```python
a.loc[:2]
```

```python
a = pd.Series(data=teman, index=['teman1','teman2','teman3','teman4'])
a.loc['teman2']
```

## Dot Notation

```python
# dengan parameter index
d = pd.Series(data=teman, index=['teman1','teman2','teman3','teman4'])
print(d['teman1'])

# dengan dot notation, kelemahannya tidak bisa menampilkan beberapa nilai sekaligus
d.teman1
```

## Boolean Mask

```python
teman = ['Joni Gudel', 'Edi Yunani', 'Toni Sulung', 'Retno Mimin']

# menyimpan series teman ke variabel c
c = pd.Series(teman)

# menampilkan teman dengan boolean mask, kelemahannya jumlah boolean harus sama dengan list
c.loc[[True, True, True, False]]
```

```python
# properti size, untuk mengetahui jumlah elemen pada series
c.size
```

```python
angka = pd.Series(range(10))

# nampilin angka 10 biji menggunakan for (karna kalo tulis True 10x kebanyakan)
angka.loc[[True for i in range(10)]]
```

```python
# pakai if untuk nampilin angka genap saja
angka.loc[[True if i%2 == 0 else False for i in range(10)]]
```

## Method Get()

```python
# gak jauh beda dengan teknik pengambilan .loc[], bedanya kalau get() gk nampilin pesan error kalo gak ada data yang ingin dicari
d = pd.Series(data=teman, index=['teman1','teman2','teman3','teman4'])

# cara biasa
print(d['teman1'])

# pakai get()
print(d.get('teman1'))
```

```python
# pakai get()
print(d.get('teman5'))
```

```python
# mengubah none dengan parameter default
print(d.get('teman5', default='data gak bisa ditemukan'))
```

```python
# manggil dengan menggunakan indeks
print(d.get(0))
```
