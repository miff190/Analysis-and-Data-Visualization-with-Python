## Usecols
untuk memilih kolom yang ingin ditampilkan

```python
namasiswa = pd.read_csv("/content/datamurid.csv", sep=';', usecols=['Nama'])
namasiswa
```

```python
# pastiin namasiswa adalah dataframe, bukan series
type(namasiswa)
```

## Squeeze
ngubah dataframe ke series

```python
namasiswa = pd.read_csv("/content/datamurid.csv", sep=';', usecols=['Nama'], squeeze=True) # <-- error, tidak kompatibel
type(namasiswa)
```

```python
namasiswa_new = pd.read_csv("/content/datamurid.csv", sep=';', usecols=['Nama'])['Nama'] # <-- solusi gak bisa pakai squeeze
type(namasiswa_new)
```

## Parameter index_col
untuk ngubah kolom tertentu menjadi indeks

```python
namasiswa = pd.read_csv("/content/datamurid.csv", sep=';', usecols=['Nama', 'IPA'], index_col='Nama')

# Mengubah kolom 'IPA' menjadi Series
namasiswa_series = namasiswa['IPA']  # <-- solusi gak bisa pakai squeeze

# cek apakah sudah jadi series
type(namasiswa_series)
```
