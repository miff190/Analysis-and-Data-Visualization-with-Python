## Membuat DataFrame

```python
import pandas as pd
```

```python
# menggunakan list
nama = ['Andi', 'Budi', 'Caca', 'Edward', 'Fina']
usia = [20, 25, 30, 42, 30]
menikah = [False, True, False, False, True]
```

```python
# masukan ke dataframe
datateman = pd.DataFrame({
    'nama' : nama,
    'usia' : usia,
    'menikah' : menikah
})
datateman
```

## Atribut loc[]
untuk mengambil data dari DataFrame

```python
# tampilan awal
namasiswa = pd.read_csv("/content/datamurid.csv", sep=';')
namasiswa
```

```python
# menggunakan atribut loc[]
namasiswa.loc[5]
```

```python
# kolom ['Nama'] jadikan indeks dulu
namasiswa = pd.read_csv("/content/datamurid.csv", sep=';', index_col='Nama')

# menggunakan kolom nama
namasiswa.loc['Donal Ceriwis']
```

```python
# menggunakan data spesifik
namasiswa.loc['Donal Ceriwis']['IPS']
```

```python
# menggunakan range
namasiswa.loc['Endang Kumpeni':'Donal Ceriwis']
```

```python
# kombinasikan range jadi lebih spesifik
namasiswa.loc[['Endang Kumpeni', 'Donal Ceriwis']]
```

```python
# kombinasikan range jadi lebih spesifik
namasiswa.loc[['Endang Kumpeni', 'Donal Ceriwis'], ['IPA','IPS']]
```
