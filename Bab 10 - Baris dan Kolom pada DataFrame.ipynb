```python
# import dataset
import pandas as pd
pemainbola = pd.read_csv('https://thinkjubilee.com/bola.csv')

# tampilan awal
pemainbola
```

## Method .iloc[]
sama kaya .loc[] tapi data dipilih berdasarkan index

```python
# mengambil 5 baris data dan 3 buah kolom
pemainbola_kecil = pemainbola.iloc[:5,0:3] # <-- :5=baris=mulai dari index 0 sampai 4; 0:3=kolom=mulai dari index 0 sampai 3
pemainbola_kecil
```

## Method .insert()
menambah kolom dan data baru menggunakan series

```python
# tampilan awal
pemainbola_kecil
```

```python
# buat series dulu
nickname = pd.Series(['Alex','Mesut','Petr','Theo','Laurent'])
nickname
```

```python
# gunakan insert untuk nambah kolom baru
pemainbola_kecil.insert(0,'nickname',nickname)
pemainbola_kecil

# ada tiga argumen dalam .insert()
# 0=index=menentukan posisi kolom (0 berarti paling kiri)
# 'nickname'=nama kolom yang diinginkan
# nickname=parameter data yang ditambahkan
```

## Method .assign()
untuk menambah kolom dan data baru dengan list

```python
# tampilan awal
pemainbola_kecil
```

```python
# tambah kolom ['goals'] posisi paling kanan
pemainbola_kecil = pemainbola_kecil.assign(goals=[15,20,18,11,14])
pemainbola_kecil
```

## Method .drop()
menghapus kolom. Argumennya berbentuk array atau list

```python
# tampilan awal
pemainbola_kecil
```

```python
# hapus kolom ['nickname']
pemainbola_kecil.drop(columns=['nickname'], inplace=True) # <-- gunakan inplace=True untuk eksekusi langsung
pemainbola_kecil
```

## Method .append()
menambah baris baru ke dataset

```python
pemainbola_kecil
```

```python
# gunakan dictionary untuk pedoman nambahin data
namajoni = pd.Series({
    'name':'Joni',
    'club': 'Juventus',
    'age': 30,
    'goals': 15
})

# terapkan dictionary ke append
pemainbola_kecil.append(namajoni)

# <-- ini bakal error karena sudah tidak berlaku lagi di pandas 2.0
# solusinnya pakai .concat() atau downgrade pandas
```

```python
# ALTERNATIF
# gunakan dictionary untuk pedoman nambahin data
namajoni = pd.Series({
    'name':'Joni Gudel',
    'club': 'Juventus',
    'age': 30,
    'goals': 15
})

# Ubah Series menjadi DataFrame
df_namajoni = pd.DataFrame([namajoni])

# menggunakan .concat()
pd.concat([pemainbola_kecil, df_namajoni])
```

```python
# gunakan ignore_index=True agar indeksnya ngurut dari 0
df_baru = pd.concat([pemainbola_kecil, df_namajoni], ignore_index=True)

# masukan gabungan df tadi ke df_baru
df_baru
```

```python
# menambah beberapa data sekaligus
namapemainbaru = pd.DataFrame({
    'name': ['Toni Sulung','Edi Yunani'],
    'club': ['Juventus','Arsenal'],
    'age': [31,28],
    'goals': [6,7]
})

# menggunakan .concat()
pd.concat([df_baru, namapemainbaru], ignore_index=True)
```
