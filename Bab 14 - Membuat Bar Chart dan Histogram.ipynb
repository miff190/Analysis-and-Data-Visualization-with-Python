## Pembuatan Bar Chart

```python
import matplotlib.pyplot as plt
Bulan = ['Jan','Feb','Mar','Apr','Mei','Jun','Jul','Ags','Sep','Okt','Nov','Des']
Penjualan = [1000,1500,1200,1300,2000,1450,1100,1000,1800,1600,1650,1100]
fig = plt.figure(figsize=(10,5))

# menggunakan method .bar()
plt.bar(Bulan, Penjualan)
plt.show()
```

```python
fig = plt.figure(figsize=(10,5))

# gunakan title dan label axis
plt.bar(Bulan, Penjualan)
plt.title('Data Penjualan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.show()
```

```python
fig = plt.figure(figsize=(10,5))

# menggunakan .subplot()
chart = plt.subplot()
chart.bar(Bulan, Penjualan, color='r')
plt.title('Data Penjualan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.show()
```

```python
fig = plt.figure(figsize=(10,5))

chart = plt.subplot()
chart.bar(Bulan, Penjualan, color='r')
plt.title('Data Penjualan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')

# gunaka method .xticks() untuk rotasi label sumbu x
plt.xticks(Bulan, rotation=90, color='b') # <-- rotation=90 artinya berputar 90 derajat
plt.show()
```

```python
fig = plt.figure(figsize=(10,5))

chart = plt.subplot()
chart.bar(Bulan, Penjualan, color='r', width=0.3) # <-- width=0.5 mengatur ketebalan chart
plt.title('Data Penjualan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')

plt.xticks(Bulan, rotation=90, color='b')
plt.show()
```

## Method .barh()
mengubah bentuk bar vertikal menjadi horisontal


```python
fig = plt.figure(figsize=(5,5))

chart = plt.subplot()
chart.barh(Bulan, Penjualan, color='r') # <-- terapkan .barh() di sini
plt.title('Data Penjualan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')

plt.yticks(Bulan, color='b') # <-- inget! kalo .barh() pake sumbu y
plt.show()
```

## Method .grid()
untuk memperlihatkan bilah garis kisi

```python
fig = plt.figure(figsize=(5,5))

chart = plt.subplot()
chart.barh(Bulan, Penjualan, color='r')
plt.title('Data Penjualan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')

plt.yticks(Bulan, color='b')
plt.grid(True) # <-- terapkan .grid() di sini
plt.show()
```

## Multiple Bar
grafik bar bertumpuk. Digunakan untuk perbandingan kelompok

```python
import matplotlib.pyplot as plt
import numpy as np

Bulan = ['Jan','Feb','Mar','Apr','Mei','Jun','Jul','Ags','Sep','Okt','Nov','Des']
Merek_ABC = [1000,1500,1200,1300,2000,1450,1100,1000,1800,1600,1650,1100]
Merek_DEF = [900,1300,2200,1900,1500,2000,1900,1600,1100,1800,2050,1500]

# numpy membantu nyiapin bar sesuai banyaknya elemen
sumbu_x = np.arange(len(Bulan))
fig = plt.figure(figsize=(10,5))

chart = plt.subplot()
chart.bar(sumbu_x - 0.2, Merek_ABC, width=0.4, label='Merek ABC', color='r')
chart.bar(sumbu_x + 0.2, Merek_DEF, width=0.4, label='Merek DEF', color='y')
plt.show()

# keterangan :
# sumbu_x - 0.2 artinya bar geser ke kiri sebanyak 0.2px
# sumbu_x + 0.2 artinya bar geser ke kanan sebanyak 0.2px
```

```python
import matplotlib.pyplot as plt
import numpy as np

Bulan = ['Jan','Feb','Mar','Apr','Mei','Jun','Jul','Ags','Sep','Okt','Nov','Des']
Merek_ABC = [1000,1500,1200,1300,2000,1450,1100,1000,1800,1600,1650,1100]
Merek_DEF = [900,1300,2200,1900,1500,2000,1900,1600,1100,1800,2050,1500]

sumbu_x = np.arange(len(Bulan))
fig = plt.figure(figsize=(10,5))

chart = plt.subplot()
chart.bar(sumbu_x - 0.2, Merek_ABC, width=0.4, label='Merek ABC', color='r')
chart.bar(sumbu_x + 0.2, Merek_DEF, width=0.4, label='Merek DEF', color='y')

# tambah title dan label
plt.title('Data Penjualan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')

plt.xticks(sumbu_x, Bulan, color='b')
plt.legend() # <-- gunakan .legend() untuk perjelas informasi
plt.show()
```

## DataFrame untuk Bar Chart

```python
import pandas as pd
import matplotlib.pyplot as plt

# membuat dataframe
nama = ['Joni','Siti','Edward','Yoyok','Mega']
sales = [30,25,27,29,30]
df_datasales = pd.DataFrame({'nama':nama, 'sales':sales})
df_datasales
```

## Histogram
untuk melihat banyak sedikitnya distribusi nilai

```python
import pandas as pd
import matplotlib.pyplot as plt

# list usia
usia = [20,24,20,50,15,
        26,13,40,50,21,
        55,45,36,41,53,
        37,37]
chart = plt.figure(figsize=(5,5))
plt.hist(usia, color='b', edgecolor='black')
plt.show()
```
