## Membuat Series

```python
# import pandas
import pandas as pd

# buat list
teman = ['Joni Gudel', 'Edi Yunani', 'Toni Sulung', 'Retno Mimin'] # [...] = list; 'Joni Gudel', ... = elemen

# ekspor ke series
pd.Series(data=teman) # pd = library pandas; .Series() = konstruktor; data = parameter; teman = argumen
```

## Data Type

```python
# tipe data int64
usia = [17, 21, 33, 45]

pd.Series(usia)
```

```python
# tipe data float64
berat = [67.5, 70.4, 73.2, 80]

pd.Series(berat)
```

```python
# tipe data object
campur = ['Mif', 70.4, 'Tah', 80]

pd.Series(campur)
```
