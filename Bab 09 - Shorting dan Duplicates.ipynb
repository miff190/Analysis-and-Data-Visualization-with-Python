## Method .sort_values()
atur urutan berdasarkan kolom tertentu

```
# This is formatted as code
```



````python
# import dataset
import pandas as pd
pemainbola = pd.read_csv('https://thinkjubilee.com/bola.csv')

# gunakan sort_values
pemainbola.sort_values(by='market_value') # <-- by=parameter untuk nentuin kolom mana yang diurut
````

````python
# gunakan ascending=False untuk urut dari terbesar ke terkecil
pemainbola.sort_values(by='market_value', ascending=False)
````

````python
# kombinasikan dengan .loc untuk memfilter data
pemainbola[
    (pemainbola['age'] < 30) &
    (pemainbola['market_value'] > 50) &
    (pemainbola['nationality'] == 'England')].sort_values(by='market_value')
````

## Method .duplicated()
untuk mencari data duplikat


````python
# tampilan umum
pemainbola.duplicated()
````

````python
# lebih detailnya
pemainbola[pemainbola.duplicated()]
````

````python
# parameter subset, untuk memilih kolom-kolom pembanding
# tujuannya untuk memastikan bahwa data yang ganda beneran sama atau beda
pemainbola[
    pemainbola.duplicated(subset=['name','club','age','nationality'])
]
````

````python
# parameter keep, mempertahankan data yang diinput pertama kali
pemainbola[
    pemainbola.duplicated(subset=['name','club','age','nationality'], keep='first')
]
````

## Method .drop_duplicates()
untuk menghapus data ganda dengan mempertahankan satu inputan aja

````python
# tampilan awal, perhatikan dimensi baris & kolomnya
pemainbola = pd.read_csv('https://thinkjubilee.com/bola.csv')
pemainbola
````

````python
# cek duplikat
pemainbola[pemainbola.duplicated()]
````

````python
# hapus dan pertahankan inputan pertama
pemainbola.drop_duplicates(keep='first')
````

## Mengatasii NaN (Not a Number)
NaN = data yang kosong

````python
pemainbola = pd.read_csv('https://thinkjubilee.com/bola.csv')

# gunakan method .isna()
pemainbola.isna()
````

````python
# pakai method .values buat ngefilter dan nampilin yang NaN aja
pemainbola[pemainbola.isna().values]
````

````python
# cari nilai rata-rata di kolom market_value
pemainbola['market_value'].mean()
````

````python
# gunakan method .fillna buat ngisi data kosong market_value dengan nilai rata-rata
pemainbola.fillna({'market_value': 11}).loc[[30,192,195]] # <-- [30,192,195] adalah no. indeks
````

````python
# gunakan method .fillna buat ngisi kolom ['posisi']
pemainbola.fillna({
    'position': 'CF',
    'market_value': 11
    }).loc[[30,192,195]]
````
