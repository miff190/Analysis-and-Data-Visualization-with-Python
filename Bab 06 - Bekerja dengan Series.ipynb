## Atribut size
unuk mengetahui banyaknya nilai pada series

```python
import pandas as pd

# read data
nilai_IPA = pd.read_csv("/content/datamurid.csv", sep=';', usecols=['Nama', 'IPA'], index_col='Nama')

# Mengubah kolom 'IPA' menjadi Series
nilai_IPA_series = nilai_IPA['IPA']

print(nilai_IPA_series)
```

```python
# size = ngitung banyaknya nilai
nilai_IPA_series.size
```

```python
# values = nampilin deretan nilai
nilai_IPA_series.values
```

```python
# index = nampilin daftar index
nilai_IPA_series.index
```

```python
# bandingin jumlah indeks dengan values
nilai_IPA_series.index.size == nilai_IPA_series.values.size
```
