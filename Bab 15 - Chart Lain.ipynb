## Violin Chart
untuk melihat distribusi data dari variabel yang berbeda

```python
import plotly.express as px
df = px.data.tips() # <-- data dummy
df
```

```python
chart = px.violin(df, x='day', y='total_bill') # <-- method yang dipakai .violin()
chart.show()

# cara membaca chart violin:
# perhatikan sumbu 'total_bill',
# 'Sun' memiliki gelembung terbesar di sekitar nilai 20
# berarti pada hari tersebut bill yang dikeluarkan jauh lebih besar dibanding nilai lainnya
```

## Box Chart
hampir sama dengan violin, bedanya di penggunaan method aja

```python
chart = px.box(df, x='day', y='total_bill') # <-- method yang dipakai .box()
chart.show()
```

## Scatter dan Buble Chart

```python
import plotly.express as px
df2 = px.data.iris() # <-- generate sample data lain
df2
```

```python
chart = px.scatter(df2, x='species', y='petal_width') # <-- method yang dipakai .scatter()
chart.show()
```

```python
# mengubahnya menjadi chart bubble dengan paramater tambahan
chart = px.scatter(df2, x='species', y='petal_width',
                   size='petal_length', color='species') # <-- parameter tambahan
chart.show()
```
