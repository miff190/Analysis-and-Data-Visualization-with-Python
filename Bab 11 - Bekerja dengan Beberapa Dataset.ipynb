```python
# persiapan url dataset
engineering = 'https://thinkjubilee.com/data/engineering.csv'
state = 'https://thinkjubilee.com/data/state.csv'
ivies = 'https://thinkjubilee.com/data/iviesleague.csv'
liberalarts = 'https://thinkjubilee.com/data/liberalarts.csv'

# read data csv
import pandas as pd
en = pd.read_csv(engineering)
st = pd.read_csv(state)
ivy = pd.read_csv(ivies)
art = pd.read_csv(liberalarts)
```

## Method .concat()
menggabung dataframes vertikal kebawah

```python
# buat variabel yang berisi list dataset
school = [en, st, ivy, art]

# gabung pakai .concat()
pd.concat(school) # <-- proses penggabungan lancar karena punya 'jumlah kolom' dan 'nama kolom' yang sama
```

## Column Axis Concatenation
penggabungan dilakukan dalam bentuk kolom

```python
# kalo gabung horisontal gunakan axis=1
pd.concat(school, axis=1) # <-- mudah dibandingkan, tapi konsekuensinya bakal banyak null
```

## Outer dan Inner
solusi penggabungan dataframe yang dimensinya beda

```python
# tampilan asli dataset 'en'. ambil 5 data teratas
en.head()
```

```python
# contoh: tambah kolom ['Location'] di dataset 'en'
en_copy = en.copy() # gunakan method .copy() agar gak recokin data aslinya
en_copy['Location'] = 'USA'
en_copy
```

```python
# gabungin data 'en_copy' dengan 'ivy' yang gak punya kolom ['Location']
pd.concat([en_copy, ivy], axis=0)
```

```python
# hal di atas serupa dengan atribut join='outer'
pd.concat([en_copy, ivy], join='outer')
```

```python
# join='inner' untuk menggabaikan kolom yang beda
pd.concat([en_copy, ivy], join='inner')
```

## Method .merge()
menggabungkan dua atau lebih DataFrame. Syarat: harus ada NAMA KOLOM YANG SAMA

```python
# buat variabel yang berisi list dataset
school = [en, st, ivy, art]
sekolah = pd.concat(school)
sekolah
```

```python
# dataframe baru
wilayah = 'https://thinkjubilee.com/data/regions.csv'
wil = pd.read_csv(wilayah)
wil
```

```python
# gabung 'sekolah' dengan dataframe baru
pd.merge(sekolah, wil)
```

### Parameter left_on dan right_on

```python
# dataset baru
income = 'https://thinkjubilee.com/data/gaji.csv'
gaji = pd.read_csv(income)
gaji
```

```python
# gabung dataset 'gaji' dengan 'sekolah'
pd.merge(sekolah, gaji) # <-- error karena gak ada nama kolom yang sama
```

```python
# solusinya gunakan attribut left_on atau right_on
pd.merge(sekolah, gaji, left_on='School Name', right_on='school_name') # <-- pastikan nama kolom sama
```

### Menggunakan method .drop()

```python
# gunakan method .drop() untuk menghapus kolom yang sama
pd.merge(sekolah, gaji, left_on='School Name', right_on='school_name').drop('school_name', axis=1) # <-- axis=1 berarti kolom
```
